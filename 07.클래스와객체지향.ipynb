#자동차 클래스 정의

- self.속성명 : 인스턴스 속성
- 속성명 : 클래스 속성

- def __init__(self): 생성자, 매개변수 => 생성자에서 인스턴스 속성 설정
- def 메소드(self): 매소드 정의

- 클래스 메소드(cls) : 클래스 속성 관리, @classmethod "데코레이터"

- 정적 메소드 : 유틸리티

```python
class Car:
    #1 - 클래스 속성
    instance_cnt = 0

    #def __init__(self): #생성자 정의
    #    print('자동차 객체가 생성되었습니다.')
    def __init__(self, color=''): #생성자 정의
        #2 - 객체의 속성
        self.color = color

        Car.instance_cnt += 1

        print(f'{self.color} 자동차 객체가 생성되었습니다.({Car.instance_cnt}대)')

    def auto_pilot(self):
        print(f'{self.color} 자동차가 자율주행중입니다.')
        print(f'{Car.instance_cnt}대의 자동차가 운행중입니다.')

    @classmethod
    def count_check(cls):
        print(f'자동차 대수 : {cls.instance_cnt}')

    @staticmethod
    def type_check(type_code):
        if type_code > 20 :
            print('전기자동차')
        elif type_code < 20 :
            print('내연자동차')
        else :
            print('하이브리드')
```

```python
class Calculator:
    #생성자
    def __init__(self, numbers):
        #인스턴스 속성 정의
        self.numbers = numbers

    #인스턴스 메소드
    def sum(self):
        return sum(self.numbers)

    def avg(self):
        return sum(self.numbers) / len(self.numbers)
```

```python
cal1 = Calculator([1,2,3,4,5])
print(cal1.sum())
print(cal1.avg())
```

```python
cal2 = Calculator([6,7,8,9,10])

print(cal2.sum()) # 40

print(cal2.avg()) #8.0
```

```python
import random

class Account:
    def __init__(self, name, balance):
        self.name = name #계좌주이름
        self.bank_name = 'SC은행'
        self.account_number = Account.makeAccount() #staticmethod 호출시 클래스로 부른다.
        self.balance = balance

    @staticmethod
    def makeAccount():
        num1 = str(random.randint(0,999)).zfill(3)
        num2 = str(random.randint(0,99)).zfill(2)
        num3 = str(random.randint(0,999999)).zfill(6)
        return f'{num1}-{num2}-{num3}'

    #인스턴스 메소드
    def displayAccount(self):
        print(f'은행명: {self.bank_name}')
        print(f'계좌번호: {self.account_number}')
        print(f'계좌주: {self.name}')
        print(f'잔액: {self.balance}')

    #입금 기능
    def deposit(self, amount):
        self.balance = self.balance + amount

    #출금 기능
    def withdraw(self, amount):
        self.balance = self.balance - amount
```

```python
acc1 = Account('smkim', 1000)
acc1.displayAccount()
```

```python
class Animal:
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return "멍멍!"

class Cat(Animal):
    def make_sound(self):
        return "야옹!"
```

```python
d1 = Dog()
d1.make_sound()
```

```python
d2 = Cat()
d2.make_sound()
```

```python
animals = [Dog(), Cat()]
for animal in animals:
    print(animal.make_sound())  # "멍멍!", "야옹!" 출력
```
