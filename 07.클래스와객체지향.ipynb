```python
class Bicycle:  #python Object 원조부모
  def __init__(self, color):
    self.color = color
  
  def move(self):
    print(f'{self.color} 자전거가 움직입니다.')
```

```python
# b1 = Bicycle() #객체생성
# b2 = Bicycle()
b1 = Bicycle('red') #객체생성
b2 = Bicycle('white')
```

```python
b1.move()
```

```python
b2.move()
```
