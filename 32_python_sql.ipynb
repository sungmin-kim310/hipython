```python
import pymysql

# 새로운 방식
con = pymysql.connections.Connection(
    host='localhost',
    user='root',
    password='1234',
    db='wntrade',
    charset='utf8'
)

# 커서 생성
cursor = con.cursor()

# 쿼리 실행 예시
sql_query = """
SELECT 사원번호, 이름, 부서명
FROM 사원 LEFT JOIN 부서
ON 사원.부서번호 = 부서.부서번호
"""

cursor.execute(sql_query)
# 쿼리 결과 가져오기
result = cursor.fetchall()
# 결과 출력
for row in result:
 print(row)
```

```python
cursor.close()
con.close()
```
