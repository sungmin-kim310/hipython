# 변수의 영역 namespace

지역변수, 전역변수

```python
def local_exam(): #사용자정의함수
    local_var = 100 # 지역변수
    print(f"함수내부: {local_var}")

local_exam()
local_var = 200 # 전역변수
local_exam()
print(f"함수외부: {local_var}")
```

```python
counter = 0
def inc_counter():
    global counter
    counter = counter + 1 # counter += 1
    print(counter)

inc_counter()
inc_counter()
inc_counter()
```
