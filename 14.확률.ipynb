```python
import random

# "이 동전은 앞면이 나올 확률이 50%야"라고 미리 정함 (Design)
expected_prob = 0.5 
trial_count = 10  # 10번 던지기
heads = 0

print(f"--- 실험 시작 (예상 확률: {expected_prob * 100}%) ---")

for i in range(trial_count):
    # 실제 동전 던지기 (0~1 사이 랜덤 숫자 뽑기)
    if random.random() < expected_prob:
        heads += 1
        result = "앞면"
    else:
        result = "뒷면"
    print(f"{i+1}번째 결과: {result}")

# --- 3. 상대도수 (Relative Frequency): 실제 결과 ---
actual_freq = heads / trial_count

print("-" * 30)
print(f"우리의 예상 (확률): {expected_prob * 100}%")
print(f"실제 결과 (상대도수): {actual_freq * 100}%")
```

```python
import math

# 1. 설정값 정의
total_balls = 7
white_balls = 4
black_balls = 3
pick_count = 3

# 2. 전체 경우의 수: C(7, 3)
total_cases = math.comb(total_balls, pick_count)

# 3. 사건의 경우의 수: C(4, 2) x C(3, 1)
# 흰공 2개 뽑기 x 검은공 1개 뽑기
event_cases = math.comb(white_balls, 2) * math.comb(black_balls, 1)

# 4. 확률 계산
probability = event_cases / total_cases

print(f"계산된 확률 (분수): {event_cases}/{total_cases}")
print(f"계산된 확률 (소수): {probability:.4f}")
```

```python
import random

def run_simulation(trials):
    act_freq = 0
    success_cnt = 0
    sample_space = ['w','w','w','w','b','b','b']

    for i in range(trials):
        drawn_balls = random.sample(sample_space, 3)
        if drawn_balls.count('w') ==2 and drawn_balls.count('b') == 1:
            success_cnt += 1
        
    act_freq = success_cnt / trials
    return act_freq
```

```python
trials = 10000
simulation_prob = run_simulation(trials)
print(f'시뮬레이션횟수 {trials}. 결과확률 {simulation_prob:.2f}')
```

```python
# 종속사건
total = 10
win = 2
prob_A = win / total # P(A)
prob_B = (win-1) / (total-1) # P(B|A)
prob_AB = prob_A * prob_B # P(A) * P(B|A)

print(f'종속사건 AB의 확률 : {prob_AB:.3f}')
```

```python
# 전체 표본공간
sample_space = [1, 2, 3, 4, 5, 6]

# 사건 A: 짝수
event_A = [x for x in sample_space if x % 2 == 0] #조건을 만족하면 리스트에 x를 담아라 

# 사건 B: 4의 약수
event_B = [x for x in sample_space if 4 % x == 0]

# 교집합 A ∩ B
event_A_and_B = list(set(event_A) & set(event_B))

# 확률 계산
P_A = len(event_A) / len(sample_space)
P_A_and_B = len(event_A_and_B) / len(sample_space)

# 조건부확률 P(B | A)
P_B_given_A = P_A_and_B / P_A

P_B_given_A
```
