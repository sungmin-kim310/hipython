````python
# 일반적인 활용
M_VALUE = 0.6214 #상수
kmh = int(input("현재 속도를 km/h 로 입력하세요. "))
if kmh > 0 :
    mph =  M_VALUE * kmh
    print(f' {kmh}를 mph로 변환하면 {mph}입니다.')
else :
    print('속도는 양수로 입력하세요')
````

3. 유저한테 두개의 float을 입력받아서, 두 수의 곱과 합을, 다음처럼 똑같이 출력하는 코드를 만드세요.
