````python
def msg(role, text):
    msg_src = {"role": role, "content" : text}
    return msg_src
````

````python
def format_gpt_output(text):
    """
    GPT 출력에서 \n을 실제 줄바꿈으로 변환하여 가독성 있게 출력
    """
    # \n을 실제 줄바꿈으로 변환
    formatted_text = text.replace('\\n', '\n')
    return formatted_text
````

````python
res = client.responses.create(
    model = 'gpt-4o-mini',
    input = [msg("system",sys_prompt), msg("user",user_prompt)],
    #input = [{"role":"system", "content":sys_prompt},
    #         {"role":"user", "content":user_prompt}],
    temperature = 0.5
)

print(format_gpt_output(res.output_text))
````

````python
def return_label(text): #text 고장내용
    response = client.responses.create(
        model="gpt-4o-mini",
        input=[
            msg("system", prompt),
            msg("user", "고장내용: " + text + "\n출력:")
        ],
        temperature=0
    )

    return ast.literal_eval(response.output_text) #출력
````

````python
def msg(role, text):
    msg_src = {"role": role, "content" : text}
    return msg_src
````

````python
def format_gpt_output(text):
    """
    GPT 출력에서 \n을 실제 줄바꿈으로 변환하여 가독성 있게 출력
    """
    # \n을 실제 줄바꿈으로 변환
    formatted_text = text.replace('\\n', '\n')
    return formatted_text
````
