# 시각화의 필요성

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
import matplotlib.pyplot as plt
```

```python
d1 = anscombe[anscombe['dataset']=='I']
d2 = anscombe[anscombe['dataset']=='II']
d3 = anscombe[anscombe['dataset']=='III']
d4 = anscombe[anscombe['dataset']=='IV']
```

```python
d1.describe()
```

```python
d2.describe()
```

```python
d3.describe()
```

```python
d4.describe()
```

```python
plt.plot(d1['x'], d1['y'], 'o')
```

```python
plt.plot(d2['x'], d2['y'], 'o')
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1) #2row, 2col,position
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot( d1['x'], d1['y'] , 'o')
axes2.plot( d2['x'], d2['y'] , 'o')
axes3.plot( d3['x'], d3['y'] , 'o')
axes4.plot( d4['x'], d4['y'] , 'o')

axes1.set_title('dataset I')
axes2.set_title('dataset II')
axes3.set_title('dataset III')
axes4.set_title('dataset IV')

fig.suptitle('Anscombe Dataset 으로 보는 데이터 시각화의 의미')
fig.set_tight_layout(True)
plt.show()
```

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6)) #차트의 크기
plt.plot(dates, prices, marker = 'o') #x축, y축, 표식

plt.subplot(1,2,2)
plt.bar(dates, prices) #x축, y축, 표식
plt.savefig('my_chart.png')
```

# 용도에 맞는 차트 그리기

```python
tips_df =sns.load_dataset('tips')
```

# 1변량 그래프 : 히스토그램 - 구간, 빈도수

```python
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8)
```

```python
# total_bill vs tip
plt.subplot(1,2,1)
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8, color='red')


plt.subplot(1,2,2)
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8)
```

# 1변량 그래프 : 박스플롯

```python
plt.boxplot(data=tips_df, x='total_bill')
plt.grid()
plt.show()
```

```python
plt.boxplot([tips_df['total_bill'], tips_df['tip']])
plt.grid()
plt.show()
```

```python
plt.boxplot(
    [tips_df[tips_df['sex']=='Female']['total_bill'],
     tips_df[tips_df['sex']=='Female']['tip']],
     labels=['total_bill/Female', 'tip/Female']
)
plt.grid()
plt.show()
```

```python
plt.boxplot(
    [tips_df[tips_df['sex']=='Male']['total_bill'],
     tips_df[tips_df['sex']=='Male']['tip']],
     labels=['total_bill/Male', 'tip/Male']
)
plt.grid()
plt.show()
```

```python
plt.boxplot(
    [tips_df[tips_df['day']=='Sun']['total_bill'],
     tips_df[tips_df['day']=='Sat']['total_bill'],
     tips_df[tips_df['day']=='Fri']['total_bill'],
     tips_df[tips_df['day']=='Thur']['total_bill']],
     labels=['total_bill/Sun', 'total_bill/Sat', 'total_bill/Fri', 'total_bill/Thur']
)
plt.grid()
plt.show()
```

```python
plt.boxplot(
    [tips_df[tips_df['time']=='Lunch']['total_bill'],
     tips_df[tips_df['time']=='Dinner']['total_bill']],
     labels=['total_bill/Lunch', 'total_bill/Dinner']
)
plt.grid()
plt.show()
```
