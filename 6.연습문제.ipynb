## ë¬¸ìì—´ ì••ì¶•í•˜ê¸°
Q2. ë¬¸ìì—´ì„ ì…ë ¥ ë°›ì•„ ê°™ì€ ë¬¸ìê°€ ì—°ì†ì ìœ¼ë¡œ ë°˜ë³µë˜ëŠ” ê²½ìš°ì— ê·¸ ë°˜ë³µ íšŸìˆ˜ë¥¼ í‘œì‹œí•´ ë¬¸ìì—´ì„ ì••ì¶•í•˜ì—¬ í‘œì‹œí•´ ë³´ì

> ì…ë ¥ì˜ˆì‹œ: aaabbcccccca

> ì¶œë ¥ì˜ˆì‹œ: a3b2c6a1

````python
#í•¨ìˆ˜ì •ì˜
def comp_str(s):
    result = ''
    if not s:
        return result

    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            result += s[i - 1] + str(count)
            count = 1

    # ë§ˆì§€ë§‰ ë¬¸ì ì²˜ë¦¬
    result += s[-1] + str(count)

    return result
````

````python
#í•¨ìˆ˜í˜¸ì¶œ
comp_str('aaabbcccccca')
````

## Duplicate Numbers
Q3. 0 ~ 9 ê¹Œì§€ì˜ ë¬¸ìë¡œ ëœ ìˆ«ìë¥¼ ì…ë ¥ë°›ì•˜ì„ ë•Œ, ì´ ì…ë ¥ê°’ì´ 0 ~ 9ê¹Œì§€ì˜ ëª¨ë“  ìˆ«ìê°€ ê°ê° í•œ ë²ˆì”©ë§Œ ì‚¬ìš©ëœ ê²ƒì¸ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•´ ë³´ì.

>ì…ë ¥ì˜ˆì‹œ : 0123456789 01234 01234567890 6789012345 012322456789

>ì¶œë ¥ì˜ˆì‹œ : true false false true false

````python
def chk_dup_num(s):
    result = True
    val_list = []
    for num in s:
        if num not in val_list:
            val_list.append(num)
        else:
            result = False
    if len(val_list) != 10 :
        result = False
    return result

print(chk_dup_num('0123456789'))
print(chk_dup_num('01234'))
print(chk_dup_num('01234567890'))
print(chk_dup_num('6789012345'))
print(chk_dup_num('012322456789'))
````
