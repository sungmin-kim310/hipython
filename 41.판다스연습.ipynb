#Series

```python
import pandas as pd
```

시리즈 : 컬럼 + 인덱스 형식으로 저장, 관리, 리스트/배열과 유사한 형태
*컬럼 : 1차원 데이터의 모음 - 1개 컬럼
*컬럼들 > 데이터프레임

데이터프레임 : 컬럼들 + 인덱스, 엑셀 표(테이블) 유사한 형태

```python
s1 = pd.Series([10, 20, 30])
s1[0:2]
```

```python
col_idx = ['A', 'B', 'C']
data = [10,20,30]
s2 = pd.Series(data, index=col_idx)
s2 #판다스 데이터타입 dtype
```

```python
s2[0] #인덱싱
```

```python
s2['A':'C'] #슬라이싱, 인덱스 이름 기반 추출
```

```python
# 시리즈 생성
names = pd.Series(['김철수', '이영희', '박민수'])
ages = pd.Series([25, 28, 22])
scores = pd.Series([85, 92, 78])

# 시리즈를 딕셔너리로 묶어 데이터프레임 생성
df = pd.DataFrame({
    '이름': names,
    '나이': ages,
    '점수': scores
})
print(df)
```

# 인덱싱

[] 인덱싱

- iloc[] 행기반 인덱싱   >>> [행, 열]
- loc[] 이름기반(위치) 인덱싱 >>> [행이름, 열이름]

- : >> start:end 까지 슬라이싱
- loc[행이름:행이름]
- iloc[행번호:행번호] >>> end 행번호는 -1까지

```python
# df[['이름','age']]
df.loc[:,['이름','age']] # loc[행,열]
```

```python
df.loc['A'] #시리즈로 꺼내짐
```

```python
df.iloc[0:3] #행번호, 열
```

```python
df.loc[:,'이름']
```

```python
df.loc[:,['이름','score']]
```

```python
df.loc['C',:]
```

#행추가와 열추가

```python
# loc를 사용하여 새로운 행 추가
df.loc['H'] = ['장민호', 26, 88]
df
```

```python
# 직접 할당으로 열 추가
df['grade'] = ['A', 'B', 'A', 'A', 'B', 'C', 'A','D']
df
```

```python
# insert() 메서드로 특정 위치에 열 추가
df.insert(loc=1, column='student_id', value=['S001', 'S002', 'S003', 'S004', 'S005', 'S006', 'S007', 'S008'])
df
```

#행 삭제와 열 삭제

-drop() 메서드를 사용하고, axis=0 매개변수를 사용하여 행을 삭제한다.

-inplace=True를 설정하면 원본 데이터프레임을 직접 수정할 수 있다

```python
# 인덱스로 행 삭제
df.drop('A', axis=0)  # 'A' 인덱스의 행 삭제

# 여러 행 한 번에 삭제
df.drop(['B', 'C'], axis=0)
df
```

```python
#axis=1 매개변수를 설정하여 열을 삭제한다. 


# 열 이름으로 열 삭제
df.drop('grade', axis=1)  # 'grade' 열 삭제

# 여러 열 한 번에 삭제
df.drop(['age', 'score'], axis=1)
```
