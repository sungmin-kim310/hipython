>> 휴대전화 번호 입력: 011-345-1922

```python
call_num = input("당신의 휴대전화 번호를 입력하세요(예:011-234-4567):")

if "011" in call_num :
    print("SKT")
elif "016" in call_num :
    print("KT")
elif "019" in call_num :
    print("LGU")
else:
    print("알수없음")
```

```python
call_num = input("당신의 휴대전화 번호를 입력하세요(예:011-234-4567):").split('-')[0]

telcom = None

if call_num == "011" :
    telcom = "SKT"
elif call_num == "016" :
    telcom = "KT"
elif call_num == "019" :
    telcom = "LGU"
elif call_num == "010" :
    telcom = "알수없음"
else:
    telcom = "번호확인필요"

if telcom == "번호확인필요" :
    print(f'{telcom}')
else:
    print(f'당신은 {telcom} 통신사의 사용자입니다.')
```

아래 코드는 비트코인의 가격 정보를 딕셔너리로 가져오는 코드이다.
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']
btc 딕셔너리 안에는 시가, 종가, 최고가, 최저가 등이 저장되어 있다.
최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장",
그렇지 않은 경우 "하락장" 문자열을 출력하라.

Key Name	Description
opening_price	최근 24시간 내 시작 거래금액
closing_price	최근 24시간 내 마지막 거래금액
min_price	최근 24시간 내 최저 거래금액
max_price	최근 24시간 내 최고 거래금액

```python
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']

변동폭 = float(btc['max_price']) - float(btc['min_price'])
시가 = float(btc['opening_price'])
최고가 = float(btc['max_price'])

if (시가+변동폭) > 최고가 :
    print("상승장")
else:
    print("하락장")
```

```python
양수음수분류기 다양한 개수의 정수를 입력받아, 각 수가 양수인지 음수인지 구분하여 각각의 개수를 출력하는 코드를 작성하시오.

입력으로부터 정수리스트로 변환하는 것은 한 줄 코드로 작성
입력 받은 값 중 0은 무시
quit 를 입력하면 프로그램을 중지합니다.
    
출력예시
정수들을 입력: 입력한 숫자가 없습니다 양수: 0개, 음수: 0개, 합계: 0 정수들을 입력: 12 -56 25 98 -54 0 77 0 양수: 4개, 음수: 2개, 합계: 102
```

```python
a = input('양수, 음수 분류 및 합계에 필요한 정수 값을 입력하시오.(예: 20 -10 25 98)').split()

양수 = 0
음수 = 0
합계 = 0

for b in a:
    if b.lower() == "quit":
        print("프로그램을 종료합니다.")
        break
    num = int(b)
    합계 += num
    
    if num > 0 :
        양수 += 1
    elif num < 0:
        음수 += 1

if 양수 == 0 and 음수 == 0 :
    print("입력한 숫자가 없습니다. 양수: 0개, 음수: 0개, 합계: 0")
else:
    print(f"양수: {양수}개, 음수: {음수}개, 합계: {합계}")
```

```python
while True :
    user_input = input('정수의 목록을 공백으로 구분해서 입력하세요. 종료: "q"')
    if user_input == 'q' :
        print("프로그램을 종료합니다.")
        break
        
    str_int = user_input.split()
    numbers = []
    
    for num in str_int:
        if num != '0' :
            numbers.append(int(num))
    
    if numbers == [] :
        print('입력한 숫자가 없습니다.')
        continue

    plus_numbers = []
    minus_numbers = []
    total_numbers = 0

    for num in numbers:
        total_numbers += num
        if num > 0 :
            plus_numbers.append(num)
        elif num < 0 :
            minus_numbers.append(num)

    print(f"양수: {len(plus_numbers)}개, 음수: {len(minus_numbers)}개, 합계: {total_numbers}")
```
