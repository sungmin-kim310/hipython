```python
df.melt_new = df.melt(id_vars='religion', var_name = 'income', value_name = 'count')
df.melt_new
```

```python
df_pivot = df.melt_new.pivot(   #long data > wide data
    index='religion',
    columns='income',
    values='count'
 ).reset_index()
df_pivot
```

```python
df_pivot1 = df.melt_new.pivot_table(   #long data > wide data
    index='religion',
    columns='income',
    values='count',
    aggfunc='min'
 ).reset_index()
df_pivot1
```

```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}
sal_df = pd.DataFrame(data)
sal_df.info()
```

```python
def age_category(age):
    if age<30 :
        return '20대'
    elif age<40 :
        return '30대'
    else:
        return '40대 이상'
```

```python
sal_df['Age_category'] = sal_df['Age'].apply(age_category)
```

```python
sal_df
```

# 보너스 컬럼 20%

```python
sal_df['Bonus'] = sal_df['Salary']*0.2
sal_df
```

```python
sal_df.insert(loc=3, column='Bonus', value=sal_df['Salary']*0.2)
sal_df
```

```python
def calc_bonus(x, rate=0.2):
    return x*rate

sal_df['Bonus'] = sal_df['Salary'].apply(calc_bonus, rate=0.3)
sal_df
```

# 데이터 병합

```python
df1 = pd.read_csv('./data/concat_1.csv')
df2 = pd.read_csv('./data/concat_2.csv')
df3 = pd.read_csv('./data/concat_3.csv')
```

```python
df1, df2, df3
```

```python
# column list가 동일 > 병합 가능
total_df = pd.concat([df1, df2, df3], ignore_index=True)
total_df
```

```python
total_df.reset_index()
```

```python
add_ser = pd.Series(['a100','b100', 'c100', 'd100'])
add_ser
```

```python
t_df = pd.concat([total_df, add_ser])
t_df
```

```python
col_list = total_df.columns.to_list()
```

```python
#DataFrame으로 만든다.
empty_df = pd.DataFrame(columns = col_list)
empty_df.loc[0] = add_ser
empty_df
```

```python
add_ser.index = empty_df.columns
```

```python
to_df = pd.concat([total_df, empty_df], ignore_index=True)
to_df
```

## merge
- 키를 기준으로 합친다.
- sql에서 join

```python
#매출 데이터
sales_data = pd.DataFrame({ 'CustomerID': [101, 102, 103, 104], 'Date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-07'], 'Amount': [250, 300, 150, 400] })
sales_data
```

```python
#고객 데이터
customer_data = pd.DataFrame({ 'CustomerID': [101, 102, 103, 105], 'Name': ['Alice', 'Bob', 'Charlie', 'Eva'], 'Region': ['North', 'South', 'East', 'West'] })
customer_data
```

```python
sales_data.info()
```

```python
customer_data.info()
```

```python
pd.merge(customer_data, sales_data, on = 'CustomerID')
```

```python
pd.merge(customer_data, sales_data, on = 'CustomerID', how='right')
```

how - 병합 기준 지정
- left
- right
- inner(default)
- outer

on - 병합을 위한 key 지정
