```python
# 계산기 만들기
# 숫자+숫자
# 결과입니다. 출력

num_1 = int(input("계산할 숫자1을 입력하시오"))
num_2 = int(input("계산할 숫자2를 입력하시오"))
cal = input("숫자 두개에 필요한 연산자를 입력하시오.(+,-,/,*):")

if cal == "+" :
    print("결과입니다.", num_1 + num_2)
elif cal == "-" :
    print("결과입니다.", num_1 - num_2)
elif cal == "/" :
    print("결과입니다.", num_1 / num_2)
else:
    print("결과입니다.", num_1 * num_2)
```

```python
# 재고관리
inventory = {
    '노트북':5,
    '마우스':120,
    '키보드':0,
    '모니터':15
}
# 0 품절, <10 부족, <50 적정, >50 충분
for name, qty in inventory.items():
#qty = inventory['노트북']
    if qty > 50 :
        status = '충분'
    elif qty > 10 :
        status = '적정'
    elif qty > 0 :
        status = '부족'
    else :
        status = '품절'
    print(f'{name} : {qty} 로 {status} 상태입니다.')
```

```python
sentence = input("문장을 입력하세요: ")
words = ["파이썬", "코딩", "데이터", "AI"]

found_word = None

for word in words:
    if word in sentence:
        found_word = word
        break

if found_word:
    print(f"입력한 문장에서 '{found_word}' 단어를 찾았습니다!")
else:
    print("입력한 문장에 해당 단어가 없습니다.")
```

```python
def check_access(user_role):
    if user_role == "admin":
        return "관리자 페이지에 접근 가능합니다."
    elif user_role == "user":
        return "일반 사용자 페이지에 접근 가능합니다."
    else:
        return "접근 권한이 없습니다."

# 테스트
print(check_access("admin"))   # 관리자 페이지에 접근 가능합니다.
print(check_access("user"))    # 일반 사용자 페이지에 접근 가능합니다.
print(check_access("guest"))   # 접근 권한이 없습니다.
```

```python
# 가위바위보
import random
player = input('가위바위보 게임 (가위/바위/보 중 하나를 입력하세요): ')

computer = random.choice(['가위','바위', '보'])

result = "패" if ( (player == '가위' and computer == '바위') or\
                   (player == '바위' and computer == '보') or\
                   (player == '보' and computer == '가위')     ) else  \
         "무승부" if player == computer else \
         "승" # 변수는 한번만
print(f'{player} : {computer} -> {result}') # 패
```
