```python
# x, y 2개의 값을 받아서 +, * 연산을 합니다.
# 결과값 2개를 반환하는 계산기 함수를 만드세요.
```

```python
def cal(x,y):
    add = int(x)+int(y)
    mul = int(x)*int(y)
    return add, mul
```

```python
# 기본 매개변수, default parameter

def introduce(name, msg='지구를 지켜줍니다.'):
    print(f'우리의 {name}님!, {msg}')

introduce('아이언맨','젠틀해요')
introduce('원더우먼',)
introduce('블랙팬서','와칸다 포에버!')
introduce(msg='어벤져스어셈블', name='캡틴아메리카') #키워드 매개변수
introduce(msg='만세')
```

```python
# * 가변인자
def 계산기(*a):
    print(a)
    return sum(a)
```

```python
def 계산기(a,b,*c):
    sum_all = a+b+sum(c)
    return sum_all
```

```python
계산기(2,3,4,5)
```

```python
# ** 매개변수 딕셔너리
def print_info(name, age, city):
    print(f'{name} [{age}] {city}')

print_info('서울','홍길동',30)
```

```python
def print_info(**kwargs):
    print(f'{name} [{age}] {city}')
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="홍길동", age=25, city="서울")
```

# CRUD : 간식박스 프로그램

1. 함수 도출: CRUD 기본
2. 함수 구현
3. 문제점 도출 - 함수 리팩토링
4. 전체흐름 구현 함수

```python
# 간식박스 생성함수
def make_snackbox():
    snackbox = ['두쫀쿠']
    return snackbox
```

```python
# 간식박스 추가함수
def add_snackbox(sbox): #간식리스트
    make_msg = input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요').split(',')
    sbox.extend(make_msg)
    print(sbox)
    return sbox

# 메뉴출력
def display_menu():
    print('1. 간식추가')
    print('2. 간식박스 확인')
    print('3. 간식 제거')
    print('4. 주문확정')
    pass
    
# 전체보기
# 삭제 함수
# 수정 함수
```

```python
def order_confirm():
    make_msg = input('주문을 확정하시겠습니까? (yes/no)').lower()
    if make_msg == 'yes' :
        return True
    else :
        return False
```

```python
def display_snackbox(sbox):
    print(sbox)
```

```python
# 서비스의 흐름
snack_box = make_snackbox()
print("간식박스가 생성완료", snack_box)
while True :
    display_menu()
    menu_no = input("메뉴를 선택하세요")
    if menu_no == '1' :
        add_snackbox(snack_box)
    elif menu_no == '2' :
        display_snackbox(snack_box)
    elif menu_no == '4' :
        display_snackbox(snack_box)
        result = order_confirm()
        if result :
            break

print('고객님의 주문을 감사합니다') 
```
