```python
# 집합연산
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)  # {1, 2, 3, 4, 5, 6} (합집합)
print(A & B)  # {3, 4} (교집합)
print(A - B)  # {1, 2} (차집합)
```
