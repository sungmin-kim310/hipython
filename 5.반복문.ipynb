```python
# 문자열 순회
isUpper = False
isDigit = False
isLower = False
isSpecial = False
special_char = '!@#$%^&*()?'

password = 'a1!!'
# 비밀번호 조건 만족여부 확인
for char in password :
    if char.isupper() : # 대문자인지 확인
        isUpper = True
    elif char.isdigit() : # 숫자인지 확인
        isDigit = True
    elif char.islower() : # 소문자인지 확인
        isLower = True
    elif char in special_char : # 특수문자 확인
        isSpecial = True
    check_result = isUpper + isDigit + isLower + isSpecial
print(f"비밀번호 강도 조건 만족여부 : { ' 만족' if check_result == 4 else '불만족'}")
```

```python
# 총일수 % 7 -> calc_day
# 입력일 : 2025-10-5
# 년도 : 1 ~ 2024 년도수 * 365 | 366
# 월 : 1 ~ 9월 * 30(4,6,9,11) | 31(1,3,5,7,8,10,12) | 28|29(2-윤년,평년)
# 일 : 일
```

```python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)


def days_in_month(year, month):
    if month in [1, 3, 5, 7, 8, 10, 12]:
        return 31
    elif month in [4, 6, 9, 11]:
        return 30
    elif month == 2:
        return 29 if is_leap_year(year) else 28


def get_weekday(year, month, day):
    total_days = 0

    # 1️⃣ 이전 연도 일수
    for y in range(1, year):
        total_days += 366 if is_leap_year(y) else 365

    # 2️⃣ 이전 월 일수
    for m in range(1, month):
        total_days += days_in_month(year, m)

    # 3️⃣ 일
    total_days += day

    # 4️⃣ 요일 계산
    week = ["월", "화", "수", "목", "금", "토", "일"]
    return week[total_days % 7]


# ===============================
# 입력 예시
year = int(input("요일을 확인할 년도를 입력하세요(예:2025): "))
month = int(input("요일을 확인할 월을 입력하세요(예:05): "))
day = int(input("요일을 확인할 일을 입력하세요(예:11): "))

print(f"{year}-{month}-{day} 요일:", get_weekday(year, month, day))
```

```python
year = int(input("요일을 확인할 년도를 입력하세요(예:2025): "))
month = int(input("요일을 확인할 월을 입력하세요(예:05): "))
day = int(input("요일을 확인할 일을 입력하세요(예:11): "))

# 연도 계산
total_days = 0

for i in range(1,year): # 2024년까지
    if i % 400 == 0: #윤년
        total_days += 366
    elif i % 100 == 0: #평년
        total_days += 365
    elif i % 4 == 0: #윤년
        total_days += 366
    else :
        total_days += 365

print(f'1st : {total_days}, {year}')

# 월 날수 계산
month_days = [0,31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
for m in range(1,month):
    total_days += month_days[m]
    
print(f'2nd : {total_days}, {year}, {month}')

# 올해가 윤년인지 체크 +1
if month >=3 :
    if year % 400 == 0: #윤년
        total_days += 1
    elif year % 100 == 0: #평년
        total_days += 0
    elif year % 4 == 0: #윤년
        total_days += 1
    else :
        total_days += 0

print(f'3nd : {total_days}, {year}, {month},{day}')

total_days += day

print(f'final : {total_days}, {year}, {month}, {day}')

# 연도 + 월 + 일 % 7 > 요일에 나타나는 수
remain = total_days % 7
result = None
요일 = ["일", "월", "화", "수", "목", "금", "토"]

result = 요일[remain]

print(f'{year}년 {month}월 {day}일은 {result}요일입니다')
```

```python
# 학생별 총점, 평균, 순위 -> 속성 추가
# 전체 총점, 평균 출력
# 과목별 평균 출력
students = [{'num':'1', 'name':'kim', 'kor':80, 'math':90,'eng':100,},
            {'num':'2', 'name':'park', 'kor':90, 'math':60,'eng':80,},
            {'num':'3', 'name':'hong', 'kor':70, 'math':70,'eng':80,},
           {'num':'4', 'name':'hong1', 'kor':70, 'math':70,'eng':80,},
           {'num':'5', 'name':'hong2', 'kor':70, 'math':70,'eng':80,},]
```

```python
for student in students:
    student['total'] = student['kor'] + student['math'] + student['eng']
    student['ave'] = round(student['total']/3,2)

for s in students:
    s['rank'] = 1

for i in range(len(students)):
    for j in range(len(students)):
        if students[i]['total'] < students[j]['total']:
            students[i]['rank'] += 1

total_sum = 0
for t in students:
    total_sum += t['total']

total_ave = round(total_sum / (len(students)*3),2)

total_kor = 0
total_math = 0
total_eng = 0

for q in students:
    total_kor = q['kor']
    total_math = q['math']
    total_eng = q['eng']

ave_kor = round(total_kor,2)
ave_math = round(total_math,2)
ave_eng = round(total_eng,2)


print(students)
print(f'전체총점 : {total_sum} \n전체평균 : {total_ave} \n국어평균 : {ave_kor} \n수학평균 : {ave_math} \n영어평균 : {ave_eng}')
```

# 업앤다운 게임

```python
import random
number = random.randint(1,10)
number # 컴퓨터의 선택
```

```python
# 정답 맞추기
cnt = 0
while True:
    answer = int(input('1~10 사이의 숫자를 입력하세요.'))
    cnt += 1
    if answer == number :
        print(f'축하합니다. 정답입니다. 도전횟수 : {cnt}')
        break
    elif answer > number :
        print('DOWN')
    elif answer < number :
        print('UP')
```
