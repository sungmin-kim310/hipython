````python
import os
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
client

# openai와 연결 > 채널
````

````python
#헬퍼
def msg(role, text) :
    msg_src = {"role" : role, "content" : [{"type":"input_text", "text":text}]}
    return msg_src
````

````python
- system
- user 
- assistant : 그전에 기억하고 싶은 대화를 저장 -> 래그로, 사람이 검수, 중간산출물
````

````python
# 프롬프트 호출 > 응답
res = client.responses.create(
    model = 'gpt-4o-mini',
    input = [
        msg("system","You must respond in english"),
        msg("user", "두바이 쫀득 쿠키 레시피")
        #{"role" : "system", "content" : "너는 취업 전문가야"}, - 헬퍼 없을 경우
        #{"role" : "user", "content" : "광고 문구를 작성"} - 헬퍼 없을 경우
    ],
    temperature=0
)
````

````python
res
````

````python
res.output_text
````

````python
esponse = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        {
            "role": "assistant", #헬퍼는 그대로 두고, assistant만 명시적으로 처리
            "content": [
                {
                    "type": "output_text",
                    "text": "바로 예상을 뚫고 우승한 브라질입니다."
                }
            ]
        },
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
````

````python
esponse.output_text
````

````python
# 헬퍼 있는 경우
def msg_assist(role, text) :
    msg_src = {"role": role, "content": [{"type":"output_text", "text":text}] }
    return msg_src
````

````python

response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        msg_assist("assistant","바로 예상을 뚫고 우승한 브라질입니다." ),
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
response.output_text
````

````python
# 파라미터의 제어

- temperature : 창의성 0~1 (최대)
- max_output_tokens : 최대 응답 길이
- top_p : 후보선정 기준 설정
- frequency_penalty : 반복 억제
````

````python
sys_prompt = """
역할 : 너는 B2B 마케팅 전문가야
목표 : 인스타그램 랜딩페이지 상단에 넣을 핵심 문구를 작성한다.
조건 : 
1. 각 문구는 20자 이내로 작성
2. 신뢰감과 전문성을 강조
3. 과장된 표현과 감탄사 금지
출력형식 : 번호목록 (1~5)
"""
````

````python
summary_prompt = """
역할 : 당신은 뉴스기사 요약의 전문가
목표 : 
1. 뉴스에 대한 핵심 주제를 작성한다.
2. 기사 내용을 5줄로 요약한다.
3. 이 기사를 읽고 독자들이나 관련 기관, 전문가들이 물어볼 수 있는 질문 3가지를 제시한다.
조건 : 
1. 한국어로 작성
2. 기사에 대한 배경지식을 가지고 있고, 전문성을 가지고 있음
3. 정제된 언어 사용
4. 전문적인 용어 사용시 하단에 용어 해석(주석: ~~ 형식)
출력형식:
주제 :  
요약 : 
가능한 질문: 
"""
````
